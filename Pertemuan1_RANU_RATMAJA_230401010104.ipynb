## Import Library & Load Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (9, 5)

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)

# tampilkan data
df.head()
```

## Statistik Deskriptif

```python
df.describe()
```

```python
print("Jumlah data per spesies:")
df["species"].value_counts()
```

## 📊 Grafik 1 — Distribusi Sepal Length

Histogram ini menunjukkan sebaran panjang sepal untuk setiap spesies iris.

```python
fig = plt.figure(figsize=(9,5))
for sp,c in zip(df["species"].cat.categories,["#4c72b0","#55a868","#c44e52"]):  
    sub = df[df["species"]==sp]
    plt.hist(sub["sepal length (cm)"],alpha=0.7,label=sp,color=c,bins=15,edgecolor="white")
plt.title("Distribusi Sepal Length per Spesies",fontsize=14,weight="bold")
plt.xlabel("Sepal Length (cm)"); plt.ylabel("Frekuensi"); plt.legend()
plt.tight_layout()
plt.show()
```

**Insight:** Spesies *setosa* memiliki sepal length paling pendek (4.3–5.8 cm), sedangkan *virginica* paling panjang (6.0–7.9 cm).

## 📈 Grafik 2 — Scatter Plot Sepal

Visualisasi hubungan antara sepal length dan sepal width per spesies.

```python
fig, ax = plt.subplots(figsize=(9,5))
for sp,c in zip(df["species"].cat.categories,["#4c72b0","#55a868","#c44e52"]):  
    sub = df[df["species"]==sp]
    ax.scatter(sub["sepal length (cm)"],sub["sepal width (cm)"],
              label=sp,color=c,alpha=0.8,s=70,edgecolors="white")
ax.set_title("Sepal Length vs Sepal Width per Spesies",fontsize=14,weight="bold")
ax.set_xlabel("Sepal Length (cm)"); ax.set_ylabel("Sepal Width (cm)"); ax.legend()
plt.tight_layout()
plt.show()
```
