## 5. Fungsi (Function)

```python
def perkenalan(nama, nim, prodi, angkatan):
    return f'{nama} | NIM: {nim} | Prodi: {prodi} | Angkatan: {angkatan}'

def hitung_rata_nilai(nilai_list):
    return sum(nilai_list) / len(nilai_list)

def grade(rata):
    if rata >= 85: return 'A'
    elif rata >= 70: return 'B'
    elif rata >= 60: return 'C'
    else: return 'D'

print(perkenalan('RANU RATMAJA', 230401010104, 'Sistem Informasi', 2023))
nilai = [85, 90, 88, 87]
rata = hitung_rata_nilai(nilai)
print(f'Rata-rata nilai: {rata}')
print(f'Grade: {grade(rata)}')
```
