## Import Library & Load Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (9, 5)

# Load dataset Titanic
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# tampilkan data
df.head()
```

## Statistik Deskriptif

```python
df.describe()
```

## Missing Values

```python
print("Missing values per kolom:")
df.isnull().sum().to_frame("Missing Values")
```

```python
print(f"Survival rate: {df['Survived'].mean():.3f}")
print(f"Rata-rata usia: {df['Age'].mean():.1f} tahun")
```

## 📊 Grafik 1 — Survival Rate per Passenger Class

Visualisasi ini menunjukkan jumlah penumpang yang selamat dan tidak selamat berdasarkan kelas penumpang.

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.countplot(data=df, x="Pclass", hue="Survived",
             palette=["#c44e52","#55a868"], ax=ax)
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles, ["Not Survived", "Survived"])
ax.set_title("Survival Rate per Passenger Class", fontsize=14, weight="bold")
ax.set_xlabel("Passenger Class"); ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```

**Insight:** Penumpang kelas 1 (First Class) memiliki tingkat keselamatan tertinggi dibandingkan kelas 2 dan 3.

**Now what?**
Kelas sosial berpengaruh signifikan terhadap peluang selamat, kemungkinan karena akses prioritas ke sekoci.

## 📈 Grafik 2 — Distribusi Umur Penumpang

Histogram ini menunjukkan sebaran umur seluruh penumpang Titanic.

```python
fig, ax = plt.subplots(figsize=(9,5))
df["Age"].dropna().hist(ax=ax, bins=30, color="#4c72b0",
                        edgecolor="white", alpha=0.85)
ax.set_title("Distribusi Umur Penumpang Titanic", fontsize=14, weight="bold")
ax.set_xlabel("Age"); ax.set_ylabel("Frequency")
plt.tight_layout()
plt.show()
```
