## Load Dataset
Dataset yang digunakan adalah **Iris Dataset** dari sklearn datasets.
Dataset ini berisi data pengukuran bunga iris dari 3 spesies berbeda.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head(5)
```

## Quiz 1
**Berapa jumlah baris dan kolom pada dataset Iris?**

```python
print('Jumlah Baris :', df.shape[0])
print('Jumlah Kolom :', df.shape[1])
print('Nama Kolom   :', df.columns)
```

## Quiz 2
**Apakah terdapat missing values pada dataset Iris?**

```python
print('Jumlah Missing Values per Kolom:')
print(df.isnull().sum())
```

## Quiz 3
**Berapa persentase jumlah bunga dari masing-masing spesies?**

```python
persen_spesies = df.groupby('species')['species'].count() / df.shape[0] * 100
print('Persentase jumlah bunga per spesies:')
print(persen_spesies)
```

## Quiz 4
**Berapa rata-rata panjang kelopak (petal_length) bunga spesies virginica?**

```python
virginica = df[df['species'] == 'virginica']
print('Rata-rata petal_length spesies virginica :', virginica['petal_length'].mean())
```

## Quiz 5
**Spesies mana yang memiliki rata-rata sepal_length terpanjang?**

```python
avg_sepal = df.groupby('species')['sepal_length'].mean()
print('Rata-rata sepal_length per spesies:')
print(avg_sepal)
print('Spesies dengan sepal_length terpanjang:', avg_sepal.idxmax())
```

## Quiz 6
**Tampilkan statistik deskriptif (mean, std, min, max) untuk setiap kolom numerik per spesies!**

```python
stats = df.groupby('species').agg(['mean', 'std', 'min', 'max']).round(3)
print('Statistik Deskriptif per Spesies:')
print(stats.to_string())
```
