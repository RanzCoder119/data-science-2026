## 1. Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_iris
print('Library berhasil diimport!')
```

## 2. Dataset & Ukuran Pemusatan Data

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print('Dataset Iris - 5 baris pertama:')
print(df.head())
print('\n=== UKURAN PEMUSATAN DATA ===')
for col in iris.feature_names:
 print(f'\n{col}:')
 print(f' Mean   : {df[col].mean():.4f}')
 print(f' Median : {df[col].median():.4f}')
 print(f' Modus  : {df[col].mode()[0]:.4f}')
```

## 3. Ukuran Penyebaran Data

```python
print('=== UKURAN PENYEBARAN DATA ===')
for col in iris.feature_names:
 print(f'\n{col}:')
 print(f' Variance   : {df[col].var():.4f}')
 print(f' Std Dev    : {df[col].std():.4f}')
 print(f' Range      : {df[col].max() - df[col].min():.4f}')
 print(f' Q1-Q3 (IQR): {df[col].quantile(0.25):.2f} - {df[col].quantile(0.75):.2f}')
print('\nRingkasan statistik lengkap:')
print(df.describe())
```

## 4. Distribusi Data - Histogram & KDE

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
colors = ['steelblue', 'coral', 'mediumseagreen', 'orchid']
for i, (ax, col, color) in enumerate(zip(axes.flatten(), iris.feature_names, colors)):
 ax.hist(df[col], bins=20, color=color, edgecolor='black', alpha=0.6, density=True)
 df[col].plot(kind='kde', ax=ax, color='darkred', linewidth=2)
 ax.axvline(df[col].mean(), color='blue', linestyle='--', linewidth=1.5, label=f'Mean={df[col].mean():.2f}')
 ax.axvline(df[col].median(), color='green', linestyle=':', linewidth=1.5, label=f'Median={df[col].median():.2f}')
 ax.set_title(f'Distribusi: {col}')
 ax.legend(fontsize=8)
plt.suptitle('Distribusi & KDE Fitur Iris', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 5. Analisis Korelasi

```python
corr_matrix = df.drop('species', axis=1).corr()
print('Matriks Korelasi:')
print(corr_matrix.round(4))
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='RdYlGn',
 linewidths=0.5, center=0, vmin=-1, vmax=1)
plt.title('Heatmap Korelasi Fitur Iris', fontsize=13)
plt.tight_layout()
plt.show()
```

## 6. Uji Korelasi Pearson & Spearman

```python
x = df['sepal length (cm)']
y = df['petal length (cm)']
pearson_r, pearson_p = stats.pearsonr(x, y)
spearman_r, spearman_p = stats.spearmanr(x, y)
print('=== Uji Korelasi: Sepal Length vs Petal Length ===')
print(f'Pearson  r={pearson_r:.4f}  p-value={pearson_p:.6f}')
print(f'Spearman r={spearman_r:.4f}  p-value={spearman_p:.6f}')
if pearson_p < 0.05:
 print('Kesimpulan: Korelasi SIGNIFIKAN (p < 0.05)')
else:
 print('Kesimpulan: Korelasi TIDAK signifikan (p >= 0.05)')
```

## 7. Analisis Per Spesies (Group Statistics)

```python
print('Statistik per Spesies:')
print(df.groupby('species').agg(['mean','std','min','max']).round(3))

# Visualisasi boxplot per spesies
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, (ax, col) in enumerate(zip(axes.flatten(), iris.feature_names)):
 sns.boxplot(data=df, x='species', y=col, palette='Set2', ax=ax)
 ax.set_title(f'{col} per Spesies')
plt.suptitle('Analisis Per Spesies - Iris Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Kesimpulan
- Ukuran pemusatan (mean, median, modus) memberikan gambaran nilai sentral dari distribusi data.
- Ukuran penyebaran (variance, std dev, IQR) menunjukkan seberapa bervariasi data dari nilai pusatnya.
- Visualisasi histogram + KDE membantu memahami bentuk distribusi data secara intuitif.
- Korelasi Pearson menunjukkan hubungan linier; Spearman untuk hubungan non-parametrik.
- Analisis per kelompok (groupby) mengungkap perbedaan statistik yang signifikan antar spesies.
