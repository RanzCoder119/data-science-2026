## Tujuan Praktikum
Pada pertemuan ini mahasiswa mempelajari teknik **Visualisasi Data** menggunakan **Matplotlib** dan **Seaborn**. Visualisasi yang dipelajari meliputi: line chart, bar chart, histogram, pie chart, scatter plot, box plot, heatmap, dan pair plot.

## 1. Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
sns.set_style('whitegrid')
print('Library siap. Matplotlib:', plt.matplotlib.__version__)
```

## 2. Load Dataset

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print('Dataset Iris:', df.shape)
print(df.head())
```

## 3. Line Chart & Bar Chart

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Line Chart - rata-rata per spesies
mean_data = df.groupby('species')['petal length (cm)'].mean()
axes[0].plot(mean_data.index, mean_data.values, marker='o', linewidth=2.5, color='steelblue', markersize=8)
axes[0].set_title('Line Chart: Rata-rata Petal Length per Spesies')
axes[0].set_xlabel('Spesies')
axes[0].set_ylabel('Rata-rata Petal Length (cm)')
axes[0].grid(True, alpha=0.5)
# Bar Chart - jumlah per spesies
count_data = df['species'].value_counts()
axes[1].bar(count_data.index, count_data.values, color=['#e74c3c','#2ecc71','#3498db'], edgecolor='black')
axes[1].set_title('Bar Chart: Jumlah Sampel per Spesies')
axes[1].set_xlabel('Spesies')
axes[1].set_ylabel('Jumlah')
for i, v in enumerate(count_data.values):
 axes[1].text(i, v + 0.5, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. Histogram & Pie Chart

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Histogram
for species, color in zip(iris.target_names, ['#e74c3c','#2ecc71','#3498db']):
 axes[0].hist(df[df['species']==species]['sepal length (cm)'],
 bins=15, alpha=0.6, label=species, color=color, edgecolor='black')
axes[0].set_title('Histogram: Distribusi Sepal Length per Spesies')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Frekuensi')
axes[0].legend()
# Pie Chart
sizes = df['species'].value_counts()
axes[1].pie(sizes, labels=sizes.index, autopct='%1.1f%%',
 colors=['#e74c3c','#2ecc71','#3498db'], startangle=90,
 explode=(0.05,0.05,0.05))
axes[1].set_title('Pie Chart: Proporsi Spesies Iris')
plt.tight_layout()
plt.show()
```

## 5. Scatter Plot

```python
plt.figure(figsize=(9, 6))
colors = {'setosa':'#e74c3c', 'versicolor':'#2ecc71', 'virginica':'#3498db'}
for species, grp in df.groupby('species'):
 plt.scatter(grp['sepal length (cm)'], grp['petal length (cm)'],
 label=species, color=colors[species], alpha=0.7, s=60, edgecolors='black', linewidths=0.4)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.title('Scatter Plot: Sepal Length vs Petal Length per Spesies', fontsize=13)
plt.legend(title='Spesies')
plt.grid(True, alpha=0.3)
plt.show()
```

## 6. Box Plot & Violin Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.boxplot(data=df, x='species', y='petal width (cm)', palette='Set2', ax=axes[0])
axes[0].set_title('Box Plot: Petal Width per Spesies')
sns.violinplot(data=df, x='species', y='sepal width (cm)', palette='Set3', ax=axes[1], inner='quartile')
axes[1].set_title('Violin Plot: Sepal Width per Spesies')
plt.tight_layout()
plt.show()
```

## 7. Heatmap Korelasi & Pair Plot

```python
# Heatmap
plt.figure(figsize=(8, 6))
corr = df.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, fmt='.3f', cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Heatmap Korelasi Fitur Iris', fontsize=13)
plt.tight_layout()
plt.show()

# Pair Plot
pair = sns.pairplot(df, hue='species', palette={'setosa':'#e74c3c','versicolor':'#2ecc71','virginica':'#3498db'},
 diag_kind='kde', plot_kws={'alpha':0.6})
pair.fig.suptitle('Pair Plot Dataset Iris', y=1.02, fontsize=13)
plt.show()
```

## Kesimpulan
- **Line chart** cocok untuk menampilkan tren data berurutan.
- **Bar chart** efektif membandingkan nilai antar kategori.
- **Histogram** menampilkan distribusi frekuensi data numerik.
- **Scatter plot** menunjukkan hubungan/korelasi antara dua variabel numerik.
- **Box & Violin plot** memvisualisasikan distribusi data beserta outliernya.
- **Heatmap** menampilkan matriks korelasi antar variabel dengan warna.
- **Pair plot** memberikan gambaran menyeluruh hubungan antar semua fitur sekaligus.
